# 1 Simplifying the sharp interface model

## 1.1 Nondimensionalisation

We start with the dimensional model as follows: 

\begin{equation}
\hat{c}_{\hat{t}} = \hat{D}_{0} \hat{c}_{\hat{y} \hat{y}},
\end{equation}

with the following condtions on the interface $\hat{y} = \hat{h}(\hat{t})$:

\begin{equation}
\hat{h}_{\hat{t}} = - \hat{\chi} \hat{k} \hat{c}, \\
\hat{\mu} \hat{c}_{\hat{y}} = \hat{k} \hat{c} + \hat{\chi} \hat{k} \hat{c}^2.
\end{equation}

Non-dimensonalise using the natural scalings: $\hat{c} = c^{\star}c$, $\hat{h} = H h$, $\hat{y} = H y$, $\hat{t} = T t$, where $T, H$ are a priori unknown. The PDE then becomes:

\begin{equation}
c_t = \dfrac{\hat{D}_{0}T}{H^2} c_{yy}.
\end{equation}

We therefore choose the time scaling:

\begin{equation}
T = \dfrac{H^2}{\hat{D}_{0}}.
\end{equation}

The equations on the interface becomes:

\begin{equation}
c_y  = \left[ \dfrac{\hat{k} H}{\hat{\mu}} \right] \left( 1 + \gamma c \right)c, \\
h_t = - \left[ \dfrac{\hat{k} T \gamma}{H} \right] c = - \left[ \dfrac{\hat{k} H \gamma}{\hat{D}_{0}} \right] c .
\end{equation}

where $\gamma = \chi c^{\star}$.

Setting the last bracketed nondimensional group to equal one, 

\begin{equation}
c_y  = \left[ \dfrac{\hat{D}_{0}}{\hat{\mu} \gamma} \right] \left( 1 + \gamma c \right)c, \\
h_t = - c.
\end{equation}

From now on, we define $\lambda := \hat{D}_{0}/\hat{\mu} \gamma$.

## 1.2 Change of variables

Consider the change of variables from $(y,t)$ to $(\zeta, \tau)$, where $t = \tau$ and $\zeta = y / h(t)$. Then:

\begin{equation}
\dfrac{\partial}{\partial t} = \dfrac{\partial}{\partial \tau}  +  \dfrac{\partial}{\partial \zeta}  \dfrac{\partial \zeta}{\partial t}  = \dfrac{\partial}{\partial \tau}  - \dfrac{\zeta h_{\tau}}{h} \dfrac{\partial}{\partial \zeta}, \\
\dfrac{\partial}{\partial y} = \dfrac{1}{h} \dfrac{\partial}{\partial \zeta}.
\end{equation}

The PDE becomes: 

\begin{equation}
c_{\tau} - \dfrac{\zeta h_{\tau}}{h} c_{\zeta} = \dfrac{1}{h^2} c_{\zeta \zeta},
\end{equation}

with conditions on the interface $\zeta = 1$ as follows:

\begin{equation}
c_{\zeta}  = \left[ \dfrac{\hat{D}_{0}}{\hat{\mu} \gamma} \right] \left( 1 + \gamma c \right)hc, \\
h_{\tau} = - c.
\end{equation}

We can actually rewrite the PDE to only involve derivatives of $c$ (and not of $h$):

\begin{equation}
h^2 c_{\tau} + \zeta h c(1, \tau) c_{\zeta} = c_{\zeta \zeta},
\end{equation}

The condition on $\zeta = 0$ remains:

\begin{equation}
c = 1,
\end{equation}

whereas the conditions at $\tau = 0$ is:

\begin{equation}
h = 0.
\end{equation}

We also need to work out an appropriate initial condition for the cleanser concentration - see the section below for more.

# 2 Early time behaviour for the sharp interface model

In this section, we assume that $\lambda \sim O(1)$. From the governing PDE, we have:

\begin{equation}
h^2c_{\tau} - \zeta h h_{\tau} c_{\zeta} = c_{\zeta \zeta}
\end{equation}

So initially, when $h = 0$, we have $c_{\zeta \zeta} = 0$. Moreover, from the other conditions, we need $c =1$ at $\zeta = 0$ and $c_{\zeta} = 0$ at the interface $\zeta = 1$. Therefore, an appropriate initial condition to take is $c=1$ at $\tau = 0$. 

Physically, we cannot translate the condition of zero cleanser to the new coordinates, since the strip on which this condition holds has length zero. 

We rescale the variables as follows, with respect to a small parameter $\epsilon \ll 1$:

\begin{equation}
\tau = \epsilon^{\alpha}T, \\
h = \epsilon^{\beta} \bar{h}, \\
c = 1 - \epsilon^{\delta} \bar{c}.
\end{equation}

From the condition for $h$ at the interface, we get:
\begin{equation}
\bar{h}_T = - \epsilon^{\beta - \alpha} \left( 1 - \epsilon^{\delta} \bar{c} \right).
\end{equation}

Since we would like the interface to move (otherwise we will preserve the singularity for $h$ at later time), we need a non-trivial ODE at leading order. Therefore, we choose $\alpha = \beta = 1$ without loss of generality. Substituting this into the governing PDE, we obtain:

\begin{equation}
\bar{c}_{\zeta \zeta} = \epsilon \left(  \bar{c}_T - \dfrac{\zeta \bar{h}_T}{\bar{h}} \bar{c}){\zeta} \right).
\end{equation}

So to leading order, $\bar{c} = A(t) \zeta + B(t)$. Note that, since we have the boundary condition $c = 1$ at $\zeta = 0$, we require $\bar{c} = 0$ here, so $B(t) = 0$. So we only need to determine $A(t)$.

We now handle the other condition at the interface, which becomes:

\begin{equation}
\bar{c}_{\zeta} = - \lambda (1 + \gamma(1 - \epsilon^{\delta} \bar{c})) \epsilon^{1-\delta} \bar{h} (1 - \epsilon^{\delta} \bar{c}).
\end{equation}

Note that if $\delta > 1$, then the spatial derivative of $\bar{c}$ blows up and we have a contradiction. If instead $0 < \delta < 1$, we have $\bar{c}_{\zeta} = 0$ at $\zeta = 1$, i.e. $A(t) \equiv 0$, so therefore $\bar{c} \equiv 0$. However, we seek a regime in which the effect on cleanser concentration on the interface is non-negligible and the model is non-trivial, so we choose $\delta = 1$. In that case, to leading order:

\begin{equation}
\bar{c}_\zeta = - \lambda (1+ \gamma) \bar{h}.
\end{equation}

Therefore, $\bar{c} = \lambda(1+\gamma) \zeta T$. Therefore, for small time $\tau = \epsilon T$, we have:

\begin{equation}
\boxed{
h = -\tau, \\
c = 1 - \lambda (1+ \gamma) \tau \zeta.
}
\end{equation}

# 3 Solving the PDE using the Method of Lines

To decide on the kinds of differences to use, we need to determine the slope of characteristics and analyse the PDE in detail. Setting $g(\tau) = c(1, \tau)$, we can rewrite the PDE as:

\begin{equation}
\dfrac{\partial^2 c}{\partial \zeta^2} = f \left( \zeta, \tau, c, \dfrac{\partial c}{\partial \zeta}, \dfrac{\partial c}{\partial \tau} \right) = h^2(\tau) \dfrac{\partial c}{\partial \tau} + \zeta h(\tau) g(\tau) \dfrac{\partial c}{\partial \zeta}.
\end{equation}

Therefore, the PDE is parabolic in nature, so there must be one repeated characteristic, with slope $d \tau /d \zeta = 0$. Therefore, the characteristics are given by $\tau = \text{const}$.

We consider a uniformly spaced grid with $N + 1$ evenly spaced points $\{ \zeta_0, \zeta_1, ..., \zeta_{N} \}$. Here, $\zeta_0 = 0$ and $\zeta_{N} = 1$, so the grid spacing is $\Delta x = 1/N$. In other words, $\zeta_j = j \Delta x$. Next, we define $c_j(t) = c(\zeta_j, t)$, for $j \in \{0, 1, 2, ..., N \}$. We use central differences to approximate the partial derivatives:

\begin{equation}
\dfrac{\partial c}{\partial \zeta} \to \dfrac{c_{j+1}(t) - c_{j-1}(t)}{2 \Delta \zeta}, \\
\dfrac{\partial^2 c}{\partial \zeta^2} \to \dfrac{c_{j-1}(t) - 2c_j(t) + c_{j+1}(t)}{(\Delta \zeta)^2}.
\end{equation}

The reason for using the centered difference is slightly delicate: an explanation can be found here between equations $(127)$ and $(128)$: http://hplgit.github.io/prog4comp/doc/pub/p4c-sphinx-Python/._pylight006.html.

The ODEs for the method of lines then become (for $j \in \{ 1, 2, ..., N-1  \}$):

\begin{equation}
\boxed{
\dfrac{d c_j}{dt} = - \dfrac{j \Delta \zeta c_{N}}{h} \dfrac{c_{j+1}-c_{j-1}}{2 \Delta \zeta} + \dfrac{1}{h^2} \dfrac{c_{j-1} - 2c_j + c_{j+1}}{(\Delta \zeta)^2} = - \dfrac{j c_{N}}{h} \dfrac{c_{j+1}-c_{j-1}}{2} + \dfrac{1}{h^2} \dfrac{c_{j-1} - 2c_j + c_{j+1}}{(\Delta \zeta)^2}.
}
\end{equation}

To enforce the boundary conditions, we need to introduce ghost points, say the point $x_{N+1}$ for the condition on the right end. Thus, for the boundary condition at $\zeta = 1$, we have (using a centered difference for the first derivative approximation):

\begin{equation}
\dfrac{c_{N+1} - c_{N-1}}{2 \Delta \zeta} = \lambda (1 + \gamma c_{N}) h c_{N}.
\end{equation}

Thus, we define:
\begin{equation}
c_{N+1} = 2 \Delta \zeta \lambda (1 + \gamma c_{N}) h c_{N} + c_{N-1}.
\end{equation}

to be a ghost solution at the point $x_{N}$ in order to implement the boundary condition at the agent-cleanser interface.

Therefore, the ODE corresponding to $j = N$ above is now:

\begin{equation}
\boxed{
\dfrac{d c_{N}}{dt} = - \dfrac{c_{N}}{h} \lambda (1 + \gamma c_{N}) h c_{N} + \dfrac{2}{h^2} \dfrac{c_{N-1} - c_{N} +  \Delta \zeta \lambda (1 + \gamma c_{N}) h c_{N}}{(\Delta \zeta)^2}.
}
\end{equation}

The other boundary condition is $c_0 = 1$. The way to implement this is to have a trivial ODE:

\begin{equation}
\dfrac{d c_0}{dt} = 0,
\end{equation}

and also implement this value for $c_0$ into the ODE for $c_1$:

\begin{equation}
\boxed{
\dfrac{d c_1}{dt} = - \dfrac{c_{N}}{h} \dfrac{c_{2}-1}{2} + \dfrac{1}{h^2} \dfrac{1 - 2c_1 + c_{2}}{(\Delta \zeta)^2}.
}
\end{equation}

To solve the problem practically, we compile a new vector with $N+1$ entries:

\begin{equation}
\mathbf{u} = [h, c_1, c_2, ..., c_N].
\end{equation}

Then, the new problem to be solved becomes (starting the indexing with 0 as in Python, so that $c_j$ corresponds to  $u_j$ for all  $j$ and $u_0$ is just $h$):

\begin{equation}
\dfrac{d u_0}{dt} = -y_N, \\
\dfrac{d u_1}{dt} = - \dfrac{u_{N}}{u_{0}} \dfrac{u_{2}-1}{2} + \dfrac{1}{u_{0}^{2}} \dfrac{1 - 2u_1 + u_{2}}{(\Delta \zeta)^2}, \\
... \\
\dfrac{d u_j}{dt} = - \dfrac{j u_{N}}{u_0} \dfrac{u_{j+1}-u_{j-1}}{2} + \dfrac{1}{u_{0}^{2}} \dfrac{u_{j-1} - 2u_j + u_{j+1}}{(\Delta \zeta)^2}, \\
... \\
\dfrac{d u_{N}}{dt} = - \dfrac{u_{N}}{u_0} \lambda (1 + \gamma u_{N}) u_0 u_{N} + \dfrac{2}{u_{0}^{2}} \dfrac{u_{N-1} - u_{N} +  \Delta \zeta \lambda (1 + \gamma u_{N}) u_0 u_{N}}{(\Delta \zeta)^2}.
\end{equation}

Due to the early time behaviour analysis, to avoid a singularity at early time we need to initialise our numerics with the following conditions: 

\begin{equation}
h(0) = - \epsilon, \\
c(\zeta,0) = 1 - \epsilon \lambda(1 + \gamma) \zeta.
\end{equation}

In our discrete grid, this becomes:

\begin{equation}
u_0(0) = \epsilon, \\
u_j(0) = 1 - \epsilon \lambda (1+ \gamma) j \Delta \zeta,
\end{equation}

where $j \in \{1,2,3..., N \}$.

## 3.1 Coding up a MOL solver

```python
import numpy as np

def sharpinterfaceMOL(u, t, N, lmbda, gamma):
    """
    Differential equations for the sharp interface model.

    The ODEs are derived using the method of lines.
    """
    
    # dydt is the return value of this function.
    dudt = np.empty_like(u)
    
    # we generate the uniform grid of points
    dx = 1/N
    
    # Equation for h, masked as y_0
    dudt[0] = -u[N]
        
    # Equation for c_1, taking into consideration condition at c_0s
    dudt[1] = -u[N]/u[0] * (u[2]-1)/2 + (1-2*u[1]+u[2])/pow(dx*u[0],2)
    
    # Equations for c_j, where /j is in {2, 3, ..., N-2}
    
    for j in range(2, N):
        dudt[j] = -j * u[N]/u[0]* (u[j+1]-u[j-1])/2 + 1/pow(u[0],2) * (u[j+1]+u[j-1]-2*u[j])/pow(dx,2)
                
    # Equation for c_{N}, incorporating the boundary condition at x = 1  
    dudt[N] = -u[N]/u[0] * lmbda * (1 + gamma*u[N]) * u[0] * u[N] + 2/(pow(u[0]*dx,2))*(u[N-1]-u[N]+dx*lmbda*(1 + gamma*u[N])*u[0]*u[N])

    return dudt
```

## 3.2 Solving the PDE using desired parameters

```python
from scipy.integrate import odeint

# Time and space parameters

N = 200 # Number of intervals in the spatial grid on [0,1]
T = 10 # Time at stopping of simulation
time_reporting = 101 # At how many points in time do we save the solution?
x = np.linspace(0,1,N+1) # Generating the uniform grid

# Tweaking the equation parameters
eps = 0.001 # Setting the initial value h(0) = -epsilon
lmbda = 1 # value for the Breward number
gamma = 1 # this is molar volume of agent * agent concentration

# Defining the initial conditions for h and c 
h0 = -eps*np.ones(1) # Initial interface position after early time behaviour
c0 = 1 - eps*lmbda*(1+gamma)*x # Initial cleanser concentration 
u0 = np.concatenate((h0,c0[1:]))

t = np.linspace(0, T, time_reporting)

solution = odeint(sharpinterfaceMOL, u0, t, args=(N, lmbda, gamma))
```

## 3.3. Plotting the solution in the $(\zeta, c)$ plane

```python
from matplotlib import pyplot as plt # go-to plotter, MATLAB style

# Reconstructing c and h from the solution vector

c_surface = np.ones((time_reporting,1)) 
c_calculated = solution[:,1:]
c = np.hstack((c_surface,c_calculated))
h = solution[:,0]

# Plotting 

for i in range(time_reporting):
    if i % 10 == 0:
        plt.figure(0)
        plt.plot(x, c[i], label = "$t =$" + str(i/(time_reporting-1)*t[-1]))
    
plt.xlabel('Rescaled distance coordinate $(\zeta)$')
plt.ylabel('Clenser concentration $c(\zeta, t)$')
plt.legend()
plt.show()
```

## 3.4 Plotting the interface position $h(\tau)$ versus time

We actually plot $-h(\tau)$ here, because we had chosen our $h$ values to be negative and we only care about displacement from the origin.

```python
plt.figure(1)
plt.plot(t,-h)
plt.xlabel('Time $(t)$')
plt.ylabel('Position of agent-cleanser interface $h(t)$')
plt.show()
```

## 3.5 Plotting the cleanser concentration in the rescaled variables

While the problem has been fully solved, we have done so in an unnatural coordinate system, that is for a distance coordinate $\zeta$ which is fixed in the interval $[0,1]$. Since we know $h(\tau)$, we can transfer back to the original distance coordinates, namely $y = h(\tau) \zeta$. We take the negative of this, compensating for the fact that $h$ is strictly negative, to describe how cleanser concentration varies away from the surface.

```python
for i in range(time_reporting):
    if i % 10 == 0:
        plt.figure(2)
        plt.plot(-h[i]*x, c[i], label = "$t =$" + str(i/(time_reporting-1)*t[-1]))
        
plt.xlabel('Actual distance coordinate $(y)$')
plt.ylabel('Clenser concentration $c(y, t)$')
plt.legend()
plt.show()
```
